```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
# Lista de todas las tiendas
tiendas = [tienda, tienda2, tienda3, tienda4]

# Lista para guardar facturaciones
facturaciones = []

# Recorrer cada tienda
for i, t in enumerate(tiendas, start=1):
    try:
        # Verificar si existe la columna 'Precio'
        if 'Precio' in t.columns:
            # Sumar la facturación
            total = t['Precio'].sum()
            facturaciones.append(total)
            print(f"Tienda {i}: Facturación = {total}")
        else:
            print(f"Tienda {i}: No tiene columna 'Precio'.")
            facturaciones.append(0)
    except Exception as e:
        print(f"Error analizando Tienda {i}: {e}")
        facturaciones.append(0)
    finally:
        print(f"Análisis finalizado para Tienda {i}.\n")

# Crear el resumen
nombres_tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

resumen_facturacion = pd.DataFrame({
    'Tienda': nombres_tiendas,
    'Facturación': facturaciones
})

# Mostrar el resumen
print("\nResumen de Facturación:")
print(resumen_facturacion)

# Mostrar facturación total
print(f"\nFacturación total de todas las tiendas: {resumen_facturacion['Facturación'].sum()}")

```

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

# Cargar los DataFrames de las tiendas
url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

# Funciones para crear los gráficos
def crear_grafico_barras(ventas, nombre_tienda):
    plt.figure(figsize=(10, 6))
    plt.bar(ventas.index, ventas['Cantidad'])
    plt.title(f'Ventas por Categoría - {nombre_tienda} (Barras)')
    plt.xlabel('Categoría del Producto')
    plt.ylabel('Cantidad de Productos Vendidos')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

def crear_grafico_circular(ventas, nombre_tienda):
    plt.figure(figsize=(8, 8))
    plt.pie(ventas['Cantidad'], labels=ventas.index, autopct='%1.1f%%', startangle=90)
    plt.title(f'Ventas por Categoría - {nombre_tienda} (Circular)')
    plt.show()

def crear_grafico_lineas(ventas, nombre_tienda):
    plt.figure(figsize=(10, 6))
    plt.plot(ventas.index, ventas['Cantidad'], marker='o')
    plt.title(f'Ventas por Categoría - {nombre_tienda} (Líneas)')
    plt.xlabel('Categoría del Producto')
    plt.ylabel('Cantidad de Productos Vendidos')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

# Función para generar los informes por tienda
def ventas_por_categoria(tienda_data, nombre_tienda):
    if 'Categoría del Producto' in tienda_data.columns and 'Precio' in tienda_data.columns:
        ventas = tienda_data.groupby('Categoría del Producto').agg({'Precio': 'sum', 'Categoría del Producto': 'count'})
        ventas = ventas.rename(columns={'Categoría del Producto': 'Cantidad'})
        ventas['Precio'] = ventas['Precio'].apply(lambda x: "${:,.2f}".format(x))

        print(f"\nVentas por categoría para {nombre_tienda}:")
        print(ventas)

        total_productos = ventas['Cantidad'].sum()
        print(f"\nTotal de productos vendidos: {total_productos}")

        crear_grafico_barras(ventas, nombre_tienda)
        crear_grafico_circular(ventas, nombre_tienda)
        crear_grafico_lineas(ventas, nombre_tienda)

    else:
        print(f"\n{nombre_tienda} no tiene las columnas necesarias para el informe de ventas.")

# Función para generar gráficos combinados de las 4 tiendas
def graficos_combinados(tiendas):
    todas_las_tiendas = pd.concat(tiendas)
    ventas_totales = todas_las_tiendas.groupby('Categoría del Producto').agg({'Precio': 'sum', 'Categoría del Producto': 'count'})
    ventas_totales = ventas_totales.rename(columns={'Categoría del Producto': 'Cantidad'})
    ventas_totales['Precio'] = ventas_totales['Precio'].apply(lambda x: "${:,.2f}".format(x))

    crear_grafico_barras(ventas_totales, "Todas las tiendas")
    crear_grafico_circular(ventas_totales, "Todas las tiendas")
    crear_grafico_lineas(ventas_totales, "Todas las tiendas")

# Lista de todas las tiendas
tiendas = [tienda, tienda2, tienda3, tienda4]

# Generar informes por tienda
for i, t in enumerate(tiendas, start=1):
    ventas_por_categoria(t, f"Tienda {i}")

# Generar gráficos combinados
graficos_combinados(tiendas)
```

```python
# prompt: de la misma manera que se hicieron los graficos anteriores, quiero lo mismo para el promedio de calificacion de cada tienda en graficos de barras laterales. un solo grafico que muestre las cuatro tiendas

import matplotlib.pyplot as plt

def calificacion_promedio(tienda_data, nombre_tienda):
    if 'Calificación' in tienda_data.columns:
        promedio = tienda_data['Calificación'].mean()
        print(f"Calificación promedio de {nombre_tienda}: {promedio:.2f}")
        return promedio
    else:
        print(f"{nombre_tienda} no tiene la columna 'Calificación'.")
        return None

# Calcular el promedio para cada tienda
promedios = []
for i, t in enumerate(tiendas, start=1):
    promedio = calificacion_promedio(t, f"Tienda {i}")
    if promedio is not None:
        promedios.append(promedio)

# Crear el gráfico de barras
if promedios:
    nombres_tiendas = [f'Tienda {i}' for i in range(1, len(promedios) + 1)]
    plt.figure(figsize=(8, 6))
    plt.bar(nombres_tiendas, promedios)
    plt.xlabel('Tiendas')
    plt.ylabel('Calificación Promedio')
    plt.title('Calificación Promedio por Tienda')
    plt.ylim(0, 5)  # Ajustar el rango del eje y para calificaciones de 0 a 5
    plt.show()
```

# 4. Productos más y menos vendidos

```python
# prompt: quiero una tabla con los productos mas y menos vendidos de cada tienda. las columnas indican las tiendas. las filas indican el producto

import pandas as pd
def productos_mas_menos_vendidos(tiendas):
    # Crear un diccionario para almacenar los productos más y menos vendidos por tienda
    productos_extremos = {}

    for i, tienda in enumerate(tiendas, start=1):
        if 'Producto' in tienda.columns and 'Precio' in tienda.columns:
            # Agrupar por producto y sumar las cantidades vendidas
            ventas_por_producto = tienda.groupby('Producto')['Precio'].count()

            # Obtener el producto más vendido
            producto_mas_vendido = ventas_por_producto.idxmax()
            cantidad_mas_vendida = ventas_por_producto.max()

            # Obtener el producto menos vendido
            producto_menos_vendido = ventas_por_producto.idxmin()
            cantidad_menos_vendida = ventas_por_producto.min()

            productos_extremos[f'Tienda {i}'] = {
                'Más vendido': (producto_mas_vendido, cantidad_mas_vendida),
                'Menos vendido': (producto_menos_vendido, cantidad_menos_vendida)
            }
        else:
            productos_extremos[f'Tienda {i}'] = "No se puede determinar (columnas faltantes)"

    # Convertir el diccionario a un DataFrame
    df_productos_extremos = pd.DataFrame.from_dict(productos_extremos, orient='index')

    return df_productos_extremos

# Llamar a la función y mostrar el resultado
tabla_productos = productos_mas_menos_vendidos(tiendas)
tabla_productos
```
