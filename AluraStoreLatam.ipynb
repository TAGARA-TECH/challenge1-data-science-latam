```python
import pandas as pd

def agregar_unidades_vendidas(df):
    """
    Agrega una columna 'Unidades Vendidas' al DataFrame que representa el total de unidades
    vendidas por producto en la tienda. Como no existe una columna 'Unidades Vendidas',
    se calcula la cantidad de veces que se vendió cada producto.

    Args:
        df: El DataFrame de la tienda (tienda, tienda2, tienda3, tienda4).

    Returns:
        El DataFrame modificado con la columna 'Unidades Vendidas' agregada.
    """
    df['Unidades Vendidas'] = df.groupby('Producto')['Producto'].transform('count')
    return df

# Aplica la función a cada DataFrame de tienda
tienda = agregar_unidades_vendidas(tienda)
tienda2 = agregar_unidades_vendidas(tienda2)
tienda3 = agregar_unidades_vendidas(tienda3)
tienda4 = agregar_unidades_vendidas(tienda4)

# Productos más y menos vendidos para cada tienda
def obtener_productos_mas_menos_vendidos(df, nombre_tienda):
    """
    Obtiene los productos más y menos vendidos de un DataFrame de tienda.

    Args:
        df: El DataFrame de la tienda.
        nombre_tienda: El nombre de la tienda (e.g., "Tienda 1").

    Returns:
        Una tupla con dos Series: productos más vendidos y productos menos vendidos.
    """
    productos_mas_vendidos = df.groupby('Producto')['Unidades Vendidas'].sum().sort_values(ascending=False)
    productos_menos_vendidos = df.groupby('Producto')['Unidades Vendidas'].sum().sort_values(ascending=True)

    print(f"Productos más vendidos en {nombre_tienda}:\n", productos_mas_vendidos.head(10))
    print(f"\nProductos menos vendidos en {nombre_tienda}:\n", productos_menos_vendidos.head(10))
    print("-" * 30)  # Separador visual

    return productos_mas_vendidos, productos_menos_vendidos

# Obtén los resultados para cada tienda
obtener_productos_mas_menos_vendidos(tienda, "Tienda 1")
obtener_productos_mas_menos_vendidos(tienda2, "Tienda 2")
obtener_productos_mas_menos_vendidos(tienda3, "Tienda 3")
obtener_productos_mas_menos_vendidos(tienda4, "Tienda 4")
```

```python
# prompt: Lineamientos: trabajar con funciones. Modularizar.

import pandas as pd
def load_data(url):
  """Loads data from a given URL into a pandas DataFrame.

  Args:
    url: The URL of the CSV file.

  Returns:
    A pandas DataFrame containing the data.
  """
  return pd.read_csv(url)

def calculate_total_revenue(df):
  """Calculates the total revenue for a given store.

  Args:
    df: The pandas DataFrame for the store.

  Returns:
    The total revenue for the store.
  """
  df['facturacion_total'] = df['Precio'] * df['Cantidad de cuotas']
  return df['facturacion_total'].sum()

def analyze_sales_by_category(df):
  """Analyzes sales by category for a given store.

  Args:
    df: The pandas DataFrame for the store.

  Returns:
    A pandas Series with sales by category.
  """
  return df.groupby('Categoría del Producto')['Precio'].sum()

def calculate_average_rating(df):
  """Calculates the average rating for a given store.

  Args:
    df: The pandas DataFrame for the store.

  Returns:
    The average rating for the store.
  """
  return df['Calificación'].mean()

def add_units_sold(df):
    """Adds a 'Unidades Vendidas' column to the DataFrame."""
    df['Unidades Vendidas'] = df.groupby('Producto')['Producto'].transform('count')
    return df

def get_top_bottom_products(df, store_name):
    """Gets top and bottom selling products for a store."""
    top_products = df.groupby('Producto')['Unidades Vendidas'].sum().sort_values(ascending=False).head(10)
    bottom_products = df.groupby('Producto')['Unidades Vendidas'].sum().sort_values(ascending=True).head(10)
    print(f"Productos más vendidos en {store_name}:\n", top_products)
    print(f"\nProductos menos vendidos en {store_name}:\n", bottom_products)
    print("-" * 30)
    return top_products, bottom_products

# URLs of the CSV files
urls = [
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"
]

# Store names
store_names = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]

# Process each store
for i, url in enumerate(urls):
  df = load_data(url)
  total_revenue = calculate_total_revenue(df)
  sales_by_category = analyze_sales_by_category(df)
  average_rating = calculate_average_rating(df)
  df = add_units_sold(df)
  top_products, bottom_products = get_top_bottom_products(df, store_names[i])


  print(f"Facturación total de {store_names[i]}: {total_revenue}")
  print(f"Ventas por categoría en {store_names[i]}:\n{sales_by_category}")
  print(f"Calificación promedio de {store_names[i]}: {average_rating}")
```
